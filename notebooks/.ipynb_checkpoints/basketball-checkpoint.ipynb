```python
import pandas as pd
```

```python
from basketball_reference_scraper.teams import get_roster, get_team_stats, get_opp_stats, get_roster_stats, get_team_misc
```

```python
# get_team roster gives you a team's number jersey, player, position, height, weight,
# birthdate, nationality, experience, and college. get_roster(team, season_end_year)
mia_2012 = get_roster('MIA','2012')
mia_2012
```

```python
# get_team_stats gets you the stats for that season in list form. get_team_stats(team, season_end_year, data_format)
# data_format is one of 'TOTALS'|'PER_GAME'|'PER_MINUTE'|'RANK'|'YEAR/YEAR'. Default value is 'TOTALS'
mia_2012_stats = get_team_stats('MIA','2012',data_format='PER_GAME')
mia_2012_stats
```

```python
#get_opp_stats is same as above but displays oppopent stats
mia_2012_opp_stats = get_opp_stats('MIA','2012',data_format='PER_GAME')
mia_2012_opp_stats
```

```python
#get_team_misc(team, season, data_format)
# Parameters:

# team - NBA team abbreviation (e.g. 'GSW', 'SAS')
# season_end_year - Desired end year (e.g. 1988, 2011)
# data_format - One of 'TOTALS'|'RANK'. Default value is 'TOTALS'

mia_2012_misc_stats = get_team_misc('MIA','2012',data_format='TOTALS')
mia_2012_misc_stats
```
